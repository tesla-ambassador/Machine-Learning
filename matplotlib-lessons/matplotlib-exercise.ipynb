```python
car_sales = pd.read_csv('7.1 car-sales.csv')
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]', '', regex=True)
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].str[:-2].astype(int)
car_sales
```

```python
car_sales['Price'].plot()
```

```python
car_sales['Total Sales'] = car_sales['Price'].cumsum()
car_sales
```

```python
car_sales['Sale Date'] = pd.date_range('1/26/2023', periods=len(car_sales))
car_sales
```

```python
car_sales.plot('Sale Date', 'Total Sales')
```

```python
car_sales.plot('Odometer (KM)', 'Price', kind='scatter')
```

```python
heart_disease = pd.read_csv('6.1 heart-disease.csv')
heart_disease.head(10)
```

```python
heart_disease['age'].hist()
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10, 15));
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
fig, ax = plt.subplots()
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])
ax.set(title='Heart disease and cholestral',
      xlabel='Age',
      ylabel='Cholestral')
ax.axhline(over_50['chol'].mean(),
          ls='--',
          c='red');
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

fig, ax = plt.subplots()
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])
ax.set(title='Heart disease and cholestral',
      xlabel='Age',
      ylabel='Cholestral')
ax.axhline(over_50['chol'].mean(),
          ls='--',
          c='red');
ax.legend(*scatter.legend_elements(), title='Target')
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

fig, ax = plt.subplots()
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter')
ax.set(title='Heart disease and cholestral',
      xlabel='Age',
      ylabel='Cholestral')
ax.axhline(over_50['chol'].mean(),
          ls='--',
          c='red');
ax.legend(*scatter.legend_elements(), title='Target')

fig.savefig('finished-plot-excercise');
plt.subplots()
```
