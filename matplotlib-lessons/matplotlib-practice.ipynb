## Object Oriented Plotting (OOP lol!)

### Which is favorable? (pyplot vs matplotlib OO method)

* When plotting something real simple and quick, use the pyplot method
* When plotting something more advanced, use the oo method.

```python
heart_disease
```

```python
## Using pyplot
heart_disease.plot(x='age', y='chol', c='target', kind='scatter', figsize=(10, 5));
```

```python
## Using OO method mixed with pyplot
fig, ax = plt.subplots()
heart_disease.plot(kind='scatter', x='age', y='chol', c='target', ax=ax, figsize=(10, 5));
# ax.set_xlim(45, 100);
```

```python
## Using OO method alone.
## creating the data
over_50 = heart_disease[(heart_disease['age'] > 50)]

## Creating the plot
fig, ax = plt.subplots(figsize=(10, 5))

## Plotting the data
scatter = ax.scatter(x=over_50['age'], 
                     y=over_50['chol'] , 
                     c=over_50['target']);
## Customizing the plots
ax.set(xlabel='Age',
      ylabel='Cholesterol level')
## Add horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle='--')
## Add legend
ax.legend(*scatter.legend_elements(), title='Target');
```

```python
len(over_50)
```

```python
## Second plot using OO method.
fig, ax = plt.subplots(figsize = (10, 5))

scatter = ax.scatter(x=over_50['thalach'] ,
                     y=over_50['chol'] ,
                     c=over_50['target'] );
ax.set(xlabel='Thalach',
      ylabel='Cholesteral',
      title='A heart disease plot showing thalach against chol')

ax.legend(*scatter.legend_elements(), title='Target')

ax.axhline(over_50['chol'].mean(),
          ls='--');
```

## Plotting on the same axis

```python
#Plotting heart disease and age.
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                              sharex=True)
#plotting ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
#customizing ax0
ax0.set(ylabel='Cholesteral',
       title= 'Heart disease and cholesteral')
#adding legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')
#adding mean line to ax0
ax0.axhline(over_50['chol'].mean(),
           ls='--')

#plotting ax1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])
#customizing ax1
ax1.set(xlabel='Age',
       ylabel='Heart rate',
       title='Heart disease and Heart rate')
#adding legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')
#adding mean line to ax1
ax1.axhline(over_50['thalach'].mean(),
           ls='--')

#Setting overall title
fig.suptitle('Heart disease analysis', fontsize=16, fontweight='bold');
```

## How to change a style from within a style

```python
## Going to style a plot that was made using the OO method.
plt.style.use('seaborn-v0_8-whitegrid')

fig, ax = plt.subplots(figsize = (10, 5))

scatter = ax.scatter(x=over_50['thalach'] ,
                     y=over_50['chol'] ,
                     c=over_50['target'],
                    cmap='summer');
ax.set(xlabel='Thalach',
      ylabel='Cholesteral',
      title='A heart disease plot showing thalach against chol')

ax.legend(*scatter.legend_elements(), title='Target')

ax.axhline(over_50['chol'].mean(),
          ls='--');
```

```python
## Styling the heart disease plot
plt.style.use('seaborn-v0_8-whitegrid')

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                              sharex=True)
#plotting ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')
#setting the ax0 x-lim
ax0.set_xlim([50, 85])
#customizing ax0
ax0.set(ylabel='Cholesterol',
       title= 'Heart disease and cholesterol levels')
#adding legend to ax0
ax0.legend(*scatter.legend_elements(), title='Target')
#adding mean line to ax0
ax0.axhline(over_50['chol'].mean(),
           ls='--',
           c='red')

#plotting ax1
scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'],
                     cmap='winter')
#customizing ax1
ax1.set(xlabel='Age',
       ylabel='Heart rate',
       title='Heart disease and Heart rate')
#setting ax1 xlim and ylim
ax1.set_xlim([50, 85])
ax1.set_ylim([60, 200])
#adding legend to ax1
ax1.legend(*scatter.legend_elements(), title='Target')
#adding mean line to ax1
ax1.axhline(over_50['thalach'].mean(),
           ls='--',
           c='red')

#Setting overall title
fig.suptitle('Heart disease analysis', fontsize=16, fontweight='bold');
```
